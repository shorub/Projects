```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('largest financial services companies by revenue.csv')
data.head(), data.info(), data.describe()
```

```python
data.isnull().sum()
```

```python
#distribution of revenue
plt.figure(figsize=(12, 5))
sns.histplot(data['Revenue in (USD Million)'], kde = True,  color='c', bins=10)
plt.title("Distribution of Revenue")
plt.xlabel("Revenue (USD Million)")
plt.ylabel("Frequency")
plt.show()
```

```python
#Revenue across industries
plt.figure(figsize=(12, 5))
sns.barplot(x='Industry', y='Revenue in (USD Million)', data=data, palette= 'viridis')
plt.title("Revenue across Industries")
plt.xlabel("Industry")
plt.ylabel("Revenue (USD Million)")
plt.show()
```

```python
#Revenue across countries
plt.figure(figsize=(12, 5))
sns.barplot(x='Headquarters', y='Revenue in (USD Million)', data=data, palette='flare')
plt.title('Revenue across Countries')
plt.xlabel('Country')
plt.ylabel('Revenue (USD Million)')
```

```python
data_sorted = data.sort_values(by=['Revenue in (USD Million)'], ascending=False)
top_10_companies = data_sorted.head(10)
plt.figure(figsize=(15, 5))
sns.barplot(x='Company', y='Revenue in (USD Million)', data=top_10_companies, palette= 'viridis')
plt.title("Top 10 Companies by Revenue")
plt.xlabel("Companies")
plt.ylabel("Revenue (USD Million)")
plt.tick_params(axis='x', rotation=90)
plt.show()
```

```python
#Top 10 companies by Total Asset
data_sorted = data.sort_values(by=['Total Assets in (USD Millions)'], ascending=False)
top_10_companies = data_sorted.head(10)
plt.figure(figsize=(15, 5))
sns.barplot(x='Company', y='Total Assets in (USD Millions)', data=top_10_companies, palette= 'light:b')
plt.title('Top 10 Companies by Total Asset')
plt.xlabel('Companies')
plt.ylabel('Total Assets (USD Million)')
plt.tick_params(axis='x', rotation=90)
plt.show()
```

```python
#Top 10 companies by Net income
data_sorted = data.sort_values(by=['Net Income in (USD Millions)'], ascending=False)
top_10_companies = data_sorted.head(10)
plt.figure(figsize=(15, 5))
sns.barplot(x='Company', y='Net Income in (USD Millions)', data=top_10_companies, palette= 'viridis')
plt.title('Top 10 Companies by Net Income')
plt.xlabel('Companies')
plt.ylabel('Net Income (USD Million)')
plt.tick_params(axis='x', rotation=90)
plt.show()
```

```python
#Comparing average revenue, net income and assets across industries.

average_revenue = data.groupby('Industry')['Revenue in (USD Million)'].mean().reset_index()
average_net_income = data.groupby('Industry')['Net Income in (USD Millions)'].mean().reset_index()
average_assets = data.groupby('Industry')['Total Assets in (USD Millions)'].mean().reset_index()

plt.figure(figsize=(12, 5))
sns.barplot(x='Industry', y='Revenue in (USD Million)', data=average_revenue, palette='viridis', label='Average Revenue')
sns.barplot(x='Industry', y='Net Income in (USD Millions)', data=average_net_income, palette='viridis', label='Average Net Income')
sns.barplot(x='Industry', y='Total Assets in (USD Millions)', data=average_assets, palette='viridis', label='Average Assets')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Industry', y='Revenue in (USD Million)', data=data, palette='Set2')
plt.title('Revenue Distribution by Industry', fontsize=16)
plt.xticks(rotation=45, ha="right")
plt.ylabel('Revenue (USD Million)')
plt.tight_layout()
plt.show()
```

```python
#pie charts to see which industries are the biggest
revenue_by_industry = data.groupby('Industry')['Revenue in (USD Million)'].sum()
total_assests_by_industry = data.groupby('Industry')['Total Assets in (USD Millions)'].sum()
net_income_by_industry = data.groupby('Industry')['Net Income in (USD Millions)'].sum()

plt.pie(revenue_by_industry, labels=revenue_by_industry.index, autopct='%1.1f%%', startangle=90)
plt.title('Revenue by Industry')
plt.show()
```

```python
plt.pie(total_assests_by_industry, labels=total_assests_by_industry.index, autopct='%1.1f%%', startangle=90)
plt.title('Total Assets by Industry')
plt.show()
```

```python
plt.pie(net_income_by_industry, labels=net_income_by_industry.index, autopct='%1.1f%%', startangle=90)
plt.title('Net Income by Industry')
plt.show()
```

```python
headquarters_count = data['Headquarters'].value_counts()

plt.figure(figsize=(12, 6))
sns.barplot(x=headquarters_count.index, y=headquarters_count.values, palette='coolwarm')
plt.title('Number of Companies by Headquarters Location', fontsize=16)
plt.xlabel('Headquarters')
plt.ylabel('Number of Companies')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
data['Profitability (%)'] = (data['Net Income in (USD Millions)'] / data['Revenue in (USD Million)']) * 100

plt.figure(figsize=(10, 6))
sns.barplot( x='Profitability (%)', y='Company', data=data.nlargest(10,
            'Profitability (%)'), palette='Spectral')
plt.title('Top 10 Most Profitable Companies (Net Income as % of Revenue)', fontsize=16)
plt.xlabel('Profitability (%)')
plt.ylabel('Company')
plt.tight_layout()
plt.show()
```

```python
#Heatmap of correlation
correlation_matrix = data[['Revenue in (USD Million)', 'Net Income in (USD Millions)', 'Total Assets in (USD Millions)']].corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Revenue, Net Income, and Total Assests', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.pairplot(data[['Revenue in (USD Million)', 'Net Income in (USD Millions)', 'Total Assets in (USD Millions)', 'Industry']], hue='Industry', palette='viridis')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x = scaler.fit_transform(data[['Revenue in (USD Million)', 'Net Income in (USD Millions)', 'Total Assets in (USD Millions)']])
x_train, x_test, y_train, y_test = train_test_split(x, data['Profitability (%)'], test_size=0.2, random_state=42)
def evaluate_model(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    return mse, r2
results = []
models = {'Linear Regression': LinearRegression(), 'Random Forest': RandomForestRegressor(), 'Decision Tree': DecisionTreeRegressor()}
for model_name, model in models.items():
    mse, r2 = evaluate_model(model, x_train, x_test, y_train, y_test)
    print(f"{model_name} - Mean Squared Error: {mse}, R-squared: {r2}")
    results.append((model_name, mse, r2))
```

```python
#comparision of the different models
results_df = pd.DataFrame(results, columns=['Model', 'Mean Squared Error', 'R-squared'])

plt.figure(figsize=(10, 6))
sns.barplot(x='Model', y='Mean Squared Error', data=results_df, color='skyblue')
plt.title('Mean Squared Error Comparison')
plt.xlabel('Model')
plt.ylabel('Mean Squared Error')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Model', y='R-squared', data=results_df, color='salmon')
plt.title('R-squared Comparison')
plt.xlabel('Model')
plt.ylabel(R'Squared')
plt.show()
```

```python
#predicted vs actual random forest
model = RandomForestRegressor()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.5)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], 'k--', lw=2, label = 'Ideal Fit', color = 'r')
plt.xlabel('Actual Profitability (%)')
plt.ylabel('Predicted Profitability (%)')
plt.title('Actual vs. Predicted Profitability Of the Random Forest Regressor')
plt.legend()
plt.show()
```
