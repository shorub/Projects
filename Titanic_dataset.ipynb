```python
rounded_age = df['Age'].round()
plt.hist(rounded_age, bins = 10, color = 'c', edgecolor = 'black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Age')
plt.show()
#Histogram showing the distribution of ages of the passengers
```

```python
survived_ages = rounded_age[df['Survived'] == 1]
plt.hist(survived_ages, bins = 10, color = 'm', edgecolor = 'black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Age of Survived Passengers')
plt.show()
#Histogram showing the distribution of ages of the survived passengers
```

```python
pass_survived = df.groupby('Survived').count()
pass_survived
```

```python
survival_rate = df.groupby('Pclass')['Survived'].mean()
survival_rate
#The following shows the survival rate of the different passenger classes
```

```python
df['Survived'].mean()
# 38.38% of the passengers survived
```

```python
survival_rate.plot(kind='bar')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Passenger Class')
plt.show()
```

```python
survival_rate_gender = df.groupby(['Sex'])['Survived'].mean()
survival_rate_gender
#This shows what percentage of males and females that survived
```

```python
plt.pie(survival_rate_gender, labels = survival_rate_gender.index, autopct='%1.1f%%')
plt.title('Survival Rate by Gender')
plt.show()
```

```python
survival_rate_genderClass = df.groupby(['Pclass', 'Sex'])['Survived'].mean()
survival_rate_genderClass
#This shows what percentage of passengers survived in each Class by gender
```

```python
survival_rate_embarked = df.groupby(['Embarked'])['Survived'].mean()
survival_rate_embarked
#This shows what percentage of passengers survived by port of embarkation
```

```python
data = df.drop(['Name', 'Ticket', 'Cabin'], axis=1)
data
```

```python
data2 = pd.get_dummies(data, columns=['Sex', 'Embarked'])
data2
```

```python
spear_corr = data2.corr(method = 'spearman')

sns.heatmap(spear_corr, annot = True, cmap = 'coolwarm')
```

```python
pear_corr = data2.corr(method = 'pearson')

sns.heatmap(pear_corr, annot = True, cmap = 'coolwarm')
plt.figure(figsize=(20, 20))
plt.show()
```
