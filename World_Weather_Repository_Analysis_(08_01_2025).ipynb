**Unsupervised Machine Learning**

Here I will be using unsupervised machine learning to put the countries into different categories based on a few features. It will categories them depending on how extreme some of these features are.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

features = ['temperature_celsius', 'wind_kph']
x = data.loc[:, features].values
x = StandardScaler().fit_transform(x)

pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2'])

kmeans = KMeans(n_clusters=3)
kmeans.fit(principalDf)

data['principal component 1'] = principalDf['principal component 1']
data['principal component 2'] = principalDf['principal component 2']
data['Cluster'] = kmeans.labels_

labels = ['Cluster 0', 'Cluster 1', 'Cluster 2']
data
data[data['Cluster'] == 0]
```

**Monte Carlo Simulation**

Monte Carlo Simulation to calculate the probability of extreme weather conditions in each cluster.

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Filter the data for Cluster 0
cluster_0_data = data[data['Cluster'] == 0]
temperature_data = cluster_0_data['temperature_celsius']
wind_speed_data = cluster_0_data['wind_kph']

# Fit distributions to temperature and wind speed
temp_mean, temp_std = norm.fit(temperature_data)
wind_mean, wind_std = norm.fit(wind_speed_data)

# Monte Carlo simulation parameters
num_simulations = 100000

# Simulate temperature and wind speed values
simulated_temperatures = np.random.normal(temp_mean, temp_std, num_simulations)
simulated_wind_speeds = np.random.normal(wind_mean, wind_std, num_simulations)

# Define thresholds for extreme weather
extreme_temp_threshold = 40  # Temperature > 40°C
extreme_wind_threshold = 50  # Wind speed > 50 kph

# Calculate probabilities of extreme events
prob_extreme_temp = np.mean(simulated_temperatures > extreme_temp_threshold)
prob_extreme_wind = np.mean(simulated_wind_speeds > extreme_wind_threshold)

# Print results
print(f"Cluster 0 - Probability of Extreme Temperature (>40°C): {prob_extreme_temp:.2%}")
print(f"Cluster 0 - Probability of Extreme Wind Speed (>50 kph): {prob_extreme_wind:.2%}")

# Plot the simulated distributions
plt.figure(figsize=(14, 6))

# Temperature Distribution
plt.subplot(1, 2, 1)
plt.hist(simulated_temperatures, bins=50, color='skyblue', alpha=0.7, label='Simulated')
plt.axvline(extreme_temp_threshold, color='red', linestyle='--', label='Threshold')
plt.title("Simulated Temperature Distribution (Cluster 0)")
plt.xlabel("Temperature (°C)")
plt.ylabel("Frequency")
plt.legend()

# Wind Speed Distribution
plt.subplot(1, 2, 2)
plt.hist(simulated_wind_speeds, bins=50, color='orange', alpha=0.7, label='Simulated')
plt.axvline(extreme_wind_threshold, color='red', linestyle='--', label='Threshold')
plt.title("Simulated Wind Speed Distribution (Cluster 0)")
plt.xlabel("Wind Speed (kph)")
plt.ylabel("Frequency")
plt.legend()

plt.tight_layout()
plt.show()
```
